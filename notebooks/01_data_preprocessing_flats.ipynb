```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/flats.csv')
print(df.shape)
df.sample(5)
```

```python
df.duplicated().sum()
```

```python
# dropping columns {'link','property_id'}
df.drop(columns=['link','property_id'],inplace=True)
```

```python
df.shape
```

```python
# correcting column name of area, changing it to price_per_sqft
df.rename(columns={'area':'price_per_sqft'},inplace=True)
df.head()
```

```python
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
# removing ratings from the names of the society
df['society'] = df['society'].str.replace(r'\d+(\.\d+)?\s*★?', '', regex=True).str.strip()
```

```python
df['society'].value_counts()
```

```python
df['price'].value_counts()
```

```python
df[df['price'] == 'Price on Request']
```

```python
# removing those rows which have Nan and "Price on Request" in price column.
df = df.dropna(subset=['price'])
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
def convert_price(price):
    price = str(price).strip()
    if('Lac' in price):
        price = price.replace('Lac','').strip()
        return round(float(price)/100,2)
    
    elif('Crore' in price):
        price = price.replace('Crore','').strip()
        return round(float(price),2)
    else:
        pass
```

```python
df['price'] = df['price'].apply(convert_price)
```

```python
df.rename(columns={'price':'price(cr)'},inplace=True)
```

```python
df.info()
```

```python
df.sample()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.replace(r'[₹,/sq.fit.]','',regex=True).str.strip().astype(float)
```

```python
df.sample()
```

```python
df['price_per_sqft'].dtype
```

```python
df.sample(10)
```

```python
df.info()
```

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom'].isnull().sum()
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype(int)
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype(int)
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df.sample(5)
```

```python
df.fillna({'additionalRoom':'Not available'},inplace=True)
df.sample(5)
```

```python
df['floorNum'].value_counts()
```

```python
df['floorNum'].isnull().sum()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'] = (
    df['floorNum']
    .str.split(' ').str.get(0)
    .replace('Ground','0')
    .str.replace('Basement','-1')
    .str.replace('Lower','0')
    .str.replace(r'(st|nd|rd|th)','',regex=True).str.strip()
)
df.sample(5)
```

```python
df['facing'].value_counts()
```

```python
df['facing'].isnull().sum()
```

```python
df.fillna({'facing':'NA'},inplace=True)
df.sample(5)
```

```python
df.insert(loc=4,column='Area',value=round((df['price(cr)']/df['price_per_sqft'])*10000000))
df.insert(loc=2,column='property_type',value='flat')
df.sample(5)
```

```python
df.to_csv('../data/processed/flats_cleaned.csv', index=False)
```
