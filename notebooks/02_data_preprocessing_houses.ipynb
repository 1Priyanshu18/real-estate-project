```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/houses.csv')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.drop(columns=['link','property_id'],inplace=True)
```

```python
df.sample()
```

```python
df['society'].value_counts()
```

```python
df['society'] = df['society'].str.replace(r'\d+(\.\d+)?\s*★?', '', regex=True).str.strip()
```

```python
df['society'].isnull().sum()
```

```python
df.fillna({'society':'Independent'},inplace=True)
df.sample(5)
```

```python
df['price'].value_counts()
```

```python
df['price'].isnull().sum()
```

```python
df[df['price'].isnull()]
```

```python
df = df.dropna(subset=['price'])
df['price'].isnull().sum()
```

```python
df[df['price'] == 'Price on Request']
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def convert_price(price):
    price = str(price).strip()
    if('Lac' in price):
        price = price.replace('Lac','').strip()
        return round(float(price)/100,2)
    
    elif('Crore' in price):
        price = price.replace('Crore','').strip()
        return round(float(price),2)
    else:
        pass
```

```python
df['price'] = df['price'].apply(convert_price)
```

```python
df.rename(columns={'price':'price(cr)'},inplace=True)
```

```python
df.sample()
```

```python
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
df.sample()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.replace(r'(₹|/sq.ft.)','',regex=True).str.replace(',','').str.strip().astype(float)
df.sample(5)
```

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom'].isnull().sum()
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype(int)
```

```python
df.sample()
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype(int)
df.sample()
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
df.sample(5)
```

```python
df.fillna({'additionalRoom':'NA'},inplace=True)
df.sample(5)
```

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0)
df.sample(5)
```

```python
df.fillna({'facing':'NA'},inplace=True)
df.sample(5)
```

```python
df.drop(columns=['area'],inplace=True)
df.insert(4, 'Area', round((df['price(cr)'] / df['price_per_sqft']) * 10000000))
df.sample()
```

```python
df.insert(loc=2,column='property_type',value='house')
df.sample()
```

```python
df.to_csv('../data/processed/house_cleaned.csv',index=False)
```
