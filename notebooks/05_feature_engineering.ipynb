```python
import re
import pandas as pd

def extract_all_areas(text):
    if pd.isna(text):
        return {}
    
    pattern = r"(Super Built up area|Built Up area|Carpet area|Plot area)\s*[:]*\s*(\d+\.?\d*)"
    
    matches = re.findall(pattern, text, flags=re.IGNORECASE)
    
    area_dict = {}
    
    for key, value in matches:
        value = float(value)
        
        # handle sqm conversion ONLY if needed
        # if bracket exists → value already sqft → skip conversion
        if '(' not in text and 'sq.m' in text:
            value = value * 10.764
        
        area_dict[key.lower()] = value
    
    return area_dict
```

```python
def get_area(text, target):
    areas = extract_all_areas(text)
    return areas.get(target.lower(), None)


df['superBuiltUpArea'] = df['areaWithType'].apply(
    lambda x: get_area(x, 'Super Built up area')
)

df['builtUpArea'] = df['areaWithType'].apply(
    lambda x: get_area(x, 'Built Up area')
)

df['carpetArea'] = df['areaWithType'].apply(
    lambda x: get_area(x, 'Carpet area')
)

# For houses
mask = df['property_type'] == 'house'

df.loc[mask, 'carpetArea'] = df.loc[mask, 'areaWithType'].apply(
    lambda x: get_area(x, 'Plot area')
)
```

```python
def check_rooms(text,room):
    if pd.isna(text):
        return 0
    
    return 1 if room in text.lower() else 0
```

```python
df.insert(loc=14,column='poojaRoom',value=df['additionalRoom'].apply(
    lambda x: check_rooms(x,'pooja')
))
df.insert(loc=15,column='studyRoom',value=df['additionalRoom'].apply(
    lambda x: check_rooms(x,'study')
))
df.insert(loc=16,column='storeRoom',value=df['additionalRoom'].apply(
    lambda x: check_rooms(x,'store')
))
df.insert(loc=17,column='servantRoom',value=df['additionalRoom'].apply(
    lambda x: check_rooms(x,'servant')
))
df.insert(loc=18,column='others',value=df['additionalRoom'].apply(
    lambda x: check_rooms(x,'others')
))
```

3. Age Possession

```python
df['agePossession'].value_counts()
```

```python
def categorize_age_possession(value):
    if pd.isna(value):
        return "Undefined"
    if "0 to 1 Year Old" in value or "Within 6 months" in value or "Within 3 months" in value:
        return "New Property"
    if "1 to 5 Year Old" in value:
        return "Relatively New"
    if "5 to 10 Year Old" in value:
        return "Moderately Old"
    if "10+ Year Old" in value:
        return "Old Property"
    if "Under Construction" in value or "By" in value:
        return "Under Construction"
    try:
        # For entries like 'May 2024'
        int(value.split(" ")[-1])
        return "Under Construction"
    except:
        return "Undefined"
```

```python
df['agePossession'] = df['agePossession'].apply(categorize_age_possession)
```
